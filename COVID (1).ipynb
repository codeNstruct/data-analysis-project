```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

#set seaborn style
sns.set(style="whitegrid")
data=pd.read_csv(r"C:\Users\anush\OneDrive\Desktop\covid_data.csv")
data['Date']==pd.to_datetime(data['Date'])

# Create dashboard layout (2 rows x 2 columns)
fig, axes = plt.subplots(2, 2, figsize=(15, 10))
fig.suptitle("Data Visual Analysis of COVID-19 Cases", fontsize=16)

# 1. Confirmed cases over time
sns.lineplot(ax=axes[0, 0], x='Date', y='Confirmed', data=data)
axes[0, 0].set_title("Confirmed Cases Over Time")
axes[0, 0].set_xlabel("Date")
axes[0, 0].set_ylabel("Confirmed Cases")
axes[0, 0].tick_params(axis='x', rotation=90)

# 2. Recovered vs Deaths
sns.lineplot(ax=axes[0, 1], x='Date', y='Recovered', data=data, label="Recovered")
sns.lineplot(ax=axes[0, 1], x='Date', y='Deaths', data=data, label="Deaths")
axes[0, 1].set_title("Recovered vs Deaths")
axes[0, 1].set_xlabel("Date")
axes[0, 1].set_ylabel("Number of Cases")
axes[0, 1].tick_params(axis='x', rotation=90)
axes[0, 1].legend()

# 3. Daily confirmed cases (Bar chart)
sns.barplot(ax=axes[1, 0], x='Date', y='Confirmed', data=data)
axes[1, 0].set_title("Daily Confirmed Cases")
axes[1, 0].set_xlabel("Date")
axes[1, 0].set_ylabel("Confirmed Cases")
axes[1, 0].tick_params(axis='x', rotation=90)

# 4. Pie chart (Case distribution)
total_confirmed = data['Confirmed'].iloc[-1]
total_recovered = data['Recovered'].iloc[-1]
total_deaths = data['Deaths'].iloc[-1]

axes[1, 1].pie(
    [total_confirmed, total_recovered, total_deaths],
    labels=["Confirmed", "Recovered", "Deaths"],
    autopct='%1.1f%%',
    startangle=140
)
axes[1, 1].set_title("COVID-19 Case Distribution")

# Adjust layout
plt.tight_layout(rect=[0, 0, 1, 0.96])
plt.show()


```
